# Exercise 1 : Matrix Operations

```python
import numpy as np

# Create a 3x3 matrix
A = np.array([[2, 5, 3],
              [1, -2, -1],
              [1, 3, 4]])

# Determinant
det_A = np.linalg.det(A)
print("Determinant:", det_A)

# Inverse (only if determinant != 0)
if det_A != 0:
    inv_A = np.linalg.inv(A)
    print("Inverse:\n", inv_A)
else:
    print("Matrix has no inverse")
```

# Exercise 6 : Basic Hypothesis Testing


```python
# Given data
productivity_before = np.random.normal(loc=50, scale=10, size=30)
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

# Hypothesis:
# H0: Training has no effect (mean difference = 0)
# H1: Training improves productivity (mean difference > 0)

# Difference
diff = productivity_after - productivity_before

# Mean difference
mean_diff = np.mean(diff)

print("Mean improvement:", mean_diff)

# Simple test (rule-based)
if mean_diff > 0:
    print("Training likely improved productivity")
else:
    print("No improvement")
```

# Exercise 8 : Time Series Data Manipulation


```python
# Generate time series for 2023
dates = np.arange('2023-01', '2024-01', dtype='datetime64[D]')
values = np.random.rand(len(dates))

# Create DataFrame
ts = pd.DataFrame({'Date': dates, 'Value': values})
ts.set_index('Date', inplace=True)

# Slicing
print("Jan-Mar:\n", ts['2023-01':'2023-03'])
print("Apr-Jun:\n", ts['2023-04':'2023-06'])
print("Jul-Sep:\n", ts['2023-07':'2023-09'])
print("Oct-Dec:\n", ts['2023-10':'2023-12'])
```
