##   Exercise 3: Optimization and Fine-Tuning of Generative AI Models
Let’s understand how to improve the output of generative AI models through prompt engineering, model fine-tuning, and dataset curation.


1. **Prompt Engineering (Improving Prompt Specificity)**

Poor prompts are underspecified. High-quality prompts constrain style, scope, audience, format, and objectives.

Original → Optimized Prompts

* Image generation
  - “Generate an image of a futuristic city.”
  - “Generate a high-resolution, photorealistic image of a futuristic coastal city in 2150, featuring vertical skyscrapers, autonomous flying vehicles, renewable energy infrastructure, and a sunset color palette, viewed from an aerial perspective.”

* Poetry
  - “Write a poem about the future.”
  - “Write a 12-line free-verse poem about the future of humanity and artificial intelligence, with a reflective and philosophical tone, using vivid imagery and avoiding rhyming.”

* Music composition
  - “Create a song in the style of classical music.”
  - “Compose a short instrumental piece inspired by late Romantic classical music, in a minor key, using piano and string orchestra, with a slow tempo and an emotional, dramatic progression.”


---

2.  **Bias and Fairness in AI Training Data**

A biased dataset leads to systematically biased outputs.

Improving Dataset Curation

* **Source diversification**: Include news from multiple political orientations, regions, and cultural contexts.

* **Data balancing**: Ensure proportional representation of viewpoints on sensitive topics.

* **Bias auditing**: Use statistical bias detection tools and human expert review.

* **Annotation guidelines**: Label data with ideological metadata to monitor skew.

* **Continuous updating**: Periodically refresh datasets to reduce historical bias.



---

3. **Fine-Tuning for Domain-Specific Tasks (Medical Research)**

Fine-tuning in medicine requires precision, traceability, and safety.

Key Steps
1. Data Collection & Preprocessing

    * Curate peer-reviewed medical papers, clinical guidelines, and validated datasets.

    * Remove outdated, retracted, or low-quality publications.

    * Normalize terminology (ICD, MeSH, drug names).

    * De-identify sensitive patient data.

2. Transfer Learning

    * Start from a pretrained language model.

    * Apply supervised fine-tuning using domain-specific corpora.

    * Use instruction tuning with expert-written prompts and answers.

    * Optionally apply RLHF (Reinforcement Learning from Human Feedback) with medical experts.

3. Evaluation Metrics

    * Factual accuracy (expert review, citation consistency).

    * Hallucination rate.

    * Precision / Recall on medical entities.

    * Consistency with known guidelines (e.g., WHO, FDA).



---


4. **Evaluating Generative AI Performance (Objective Metrics)**

**Text Generation Metrics**

* **BLEU / ROUGE** → Similarity to reference texts.

* **Perplexity** → Language fluency and predictability.

* **BERTScore** → Semantic similarity.

* **FactScore / QA-based checks** → Factual correctness.

**Image Generation Metrics**

* **FID (Fréchet Inception Distance) **→ Visual realism.

* **IS (Inception Score)** → Image diversity and quality.

* **CLIP Score** → Image–text alignment.

These metrics complement human evaluation and reduce subjectivity.


---

5. **Controlling AI Creativity and Coherence**

Uncontrolled creativity causes hallucinations; over-control causes dull outputs.

**Techniques**

* **Temperature Scaling**

    Low temperature (≈0.2–0.5): deterministic, factual outputs.

    High temperature (>0.8): creative but risky outputs.

* **Reinforcement Learning (RLHF)**

    Reward factual, coherent, and safe responses.

    Penalize contradictions and hallucinations.

* **Attention Mechanisms**

    Improve long-range coherence.

    Maintain contextual consistency in long texts.

Optimal performance requires balanced creativity + strict constraints.
