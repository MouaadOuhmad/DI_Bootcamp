2) Discuss: How did the shift to modern NLP impact scalability and efficiency?

- TODO: Write 5-7 sentences connecting representation learning, hardware parallelism, and transfer learning.

The shift to modern NLP significantly improved scalability by replacing manual feature engineering with **representation learning**, allowing models to automatically extract hierarchical linguistic features from raw text. Contextual embeddings enable a single pre-trained model to generalize across many tasks, reducing the need to design and maintain task-specific pipelines. This paradigm aligns naturally with **hardware parallelism**, as transformer architectures are highly optimized for GPUs and TPUs, enabling efficient large-scale training through matrix operations. Although pre-training is computationally expensive, the cost is amortized over many downstream applications via **transfer learning**. Fine-tuning requires comparatively little data and compute, making adaptation to new tasks faster and more efficient than training traditional models from scratch. As a result, modern NLP systems scale better in both performance and deployment scope, despite higher upfront computational demands.



## 🌟 Exercise 2: LLM Architecture and Application Scenarios
For each, describe (a) core architectural differences, (b) a real application, (c) why it fits.

### BERT
- **Core architectural differences**
Bidirectional encoder-only Transformer, trained with Masked Language Modeling (MLM) and historically Next Sentence Prediction (NSP).
- **Application:** Named Entity Recognition (NER), sentiment analysis, and extractive question answering.
- **Why:** BERT's bidirectional attention allows each token to leverage both left and right context, which is crucial for precise semantic understanding and token-level predictions. Since these tasks require understanding rather than generation, an encoder-only architecture is computationally efficient and highly effective.

### GPT
- **Core architectural differences**
Decoder-only Transformer trained using autoregressive causal language modeling, where each token depends only on previous tokens.
- **Application:** Chatbots, open-ended text generation, and code generation.
- **Why:** GPT's causal objective directly matches the next-token prediction required in generative tasks. The unidirectional design ensures coherent sequence generation, making it ideal for interactive and creative applications where fluent text continuation is essential.

### T5
- **Core architectural differences**
Full encoder-decoder Transformer using a unified text-to-text framework, where all tasks are framed as input-text → output-text.
- **Application**: Text summarization and machine translation.
- **Why:** These tasks require transforming a complete input sequence into a new output sequence. T5’s encoder captures global input semantics, while the decoder generates structured outputs, making it particularly effective for sequence-to-sequence transformations and multi-task learning.


## 🌟 Exercise 3: Benefits and Ethics of Pre-training
- Benefits (explain each in your words):
  1) Improved generalization: Pre-trained models learn universal language patterns, allowing them to perform well on unseen tasks and data.
  2) Less labeled data: Since models are trained on large unlabeled corpora, only small labeled datasets are needed for downstream tasks.
  3) Faster fine-tuning: Pre-training provides good initial weights, reducing training time and computational cost during task adaptation.
  4) Transfer learning: Knowledge learned once can be reused across many NLP tasks, improving efficiency and scalability.
  5) Robustness: Exposure to diverse data makes models more resilient to noise, paraphrasing, and domain shifts.

- Ethical concerns: Pre-training can amplify bias, generate misinformation, enable misuse, and raise privacy risks due to large, uncontrolled datasets.
- Mitigations: Careful data curation, differential privacy, safety filters, RLHF, and continuous audits help reduce ethical risks.

## 🌟 Exercise 4: Transformer Architecture Deep Dive
### Self-Attention & Multi-Head Attention
- **Q / K / V Flow and Attention Computation**
In self-attention, each input token embedding is linearly projected into three vectors: **Query (Q)**, **Key (K)**, and **Value (V)**. For a given token, its query is compared with the keys of all tokens using a dot product to measure relevance. These scores are scaled and passed through a softmax, producing attention weights that sum to one. Each output representation is then a weighted sum of the value vectors, allowing the token to aggregate information from relevant parts of the sequence.
- **Why Multi-Head Attention Helps**
Multi-head attention applies multiple independent Q/K/V projections in parallel. Each head learns to focus on a different **representation subspace**, enabling the model to capture diverse linguistic relations simultaneously (e.g., syntax, semantics, long-range dependencies). The outputs of all heads are concatenated and linearly combined, yielding a richer contextual representation than a single attention head.

- **Example Sentence and Relations Captured**
Sentence:
*“The engineer who designed the bridge said it would last decades.”*

  - Head 1 (syntactic relation): Links “engineer” to “designed” (subject–verb relationship).

  - Head 2 (coreference/semantic relation): Connects “it” to “the bridge”, resolving pronoun reference.

### Pre-training Objectives
- MLM vs CLM:
  - **Masked Language Modeling (MLM)**: Random tokens are masked and predicted using bidirectional context. Common in encoder-based models like BERT.

  - **Causal Language Modeling (CLM)**: Tokens are predicted left-to-right, using only past context. Used in decoder-only models like GPT.
- When to prefer MLM vs CLM: MLM is preferred for understanding tasks such as classification, NER, and extractive QA, where full context is available. CLM is preferred for generation tasks, including dialogue, storytelling, and code generation, where sequential output is required.
- Next Sentence Prediction (NSP) NSP was introduced in early BERT to help the model learn sentence-to-sentence coherence for tasks like QA. However, many modern models avoid NSP because it was found to add limited value, and alternative objectives (e.g., sentence-order prediction or longer-context training) provide better discourse modeling)

### Transformer Model Selection
- **Sentiment Analysis on Reviews**
  *Encoder-only (e.g., BERT)*Justification: The task requires understanding the full input text rather than generating output. Bidirectional context improves sentiment detection.

- **Conversational Chatbot (Creative Responses)**
  *Decoder-only (e.g., GPT)*
Justification: Autoregressive generation is ideal for producing coherent, creative, and context-aware responses token by token.

- **Technical Document Translation (EN → ES)**
  *Encoder-Decoder (e.g., T5, Transformer)*
Justification: Translation requires encoding the full source sentence and generating a new structured target sequence.n

### Positional Encoding
- Purpose: Self-attention is permutation-invariant, meaning it does not inherently capture word order. Positional encoding injects absolute or relative position information so the model can distinguish between different token sequences and understand syntax and meaning.
- Example failure without positions: The sentences “The dog bit the man” and “The man bit the dog” would appear identical to the model without positional encodings, leading to incorrect or ambiguous interpretations

## 🌟 Exercise 6: Softmax Temperature: The Randomness Regulator
### 1) Temperature Scenarios
- T=0.2: Low temperature
  - Very deterministic, repetitive, high accuracy, low creativity.
- T=1.5: High temperature
  - More random and creative, but higher risk of incoherence and errors.
- T=1.0: Temperature = 1
  - Balanced diversity and coherence.

### 2) Application Design
- Bedtime stories (creativity vs coherence): Use a moderately high temperature to encourage creativity while keeping the story understandable
- Financial report summaries (accuracy/reliability): Use a low temperature to ensure factual accuracy and reliability.

### 3) Temperature & Bias
- Low temperature can reinforce dominant (biased) patterns; higher temperature increases diversity but may introduce unstable or inappropriate outputs.


### (Optional) Quick Generation Demo Across Temperatures
Note: Requires `transformers` and model download; skip if offline.


```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

model_name = "gpt2"  # TODO: choose a causal LM

prompt = "Artificial intelligence will"  # TODO: write your own prompt

pipe = pipeline(
    "text-generation",
    model=AutoModelForCausalLM.from_pretrained(model_name),
    tokenizer=AutoTokenizer.from_pretrained(model_name),
    device=0 if torch.cuda.is_available() else -1,
)

for T in [0.2, 1.0, 1.5]:
    out = pipe(prompt, max_new_tokens=40, temperature=T, do_sample=True)
    print("--- Temperature:", T)
    print(out[0]["generated_text"])  # Inspect how style changes with T
```
